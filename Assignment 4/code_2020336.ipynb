```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df1)
df_scaled = pd.DataFrame(df_scaled, columns=df1.columns)
df_scaled.head()
```

### Part B: PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
pca1 = pca.fit_transform(df_scaled)
cols = []
for i in range(1, 10):
    cols.append('PC'+str(i))
pca1 = pd.DataFrame(pca1, columns=cols)
pca1
```

```python
print(pca.explained_variance_ratio_)
sum = 0
for i in range(0,len(pca.explained_variance_ratio_)):
    sum = sum + pca.explained_variance_ratio_[i]
    print(sum)
```

```python
plt.bar(cols, pca.explained_variance_ratio_)
plt.xlabel('PCA Components')
plt.ylabel('Variance Ratio')
plt.title('PCA Variance Ratio')
plt.show()
```

```python
#Choosing the number of components = 4
pca_final = pca1[['PC1', 'PC2', 'PC3', 'PC4']]
```

```python
sns.heatmap(pca_final.corr(), annot=True, cmap='coolwarm')
```

```python
#scatter plot for all columns of pca_final
for i in pca_final.columns:
    for j in pca_final.columns:
        if i == j:
            continue
        sns.scatterplot(data = pca_final,x = pca_final[i], y = pca_final[j])
        plt.show()
```

```python
#Heat map for correlation between original data and pca_final
pca_final_corr = pd.DataFrame()
pca_final_corr['PC1'] = df_scaled.corrwith(pca_final['PC1'])
pca_final_corr['PC2'] = df_scaled.corrwith(pca_final['PC2'])
pca_final_corr['PC3'] = df_scaled.corrwith(pca_final['PC3'])
pca_final_corr['PC4'] = df_scaled.corrwith(pca_final['PC4'])
pca_final_corr.head()
sns.heatmap(pca_final_corr, annot=True, cmap='coolwarm')
plt.show()

```

```python
#Final Data to be used for k-means clustering
pca = PCA(n_components=4)
df_final = pca.fit_transform(df_scaled)
df_final = pd.DataFrame(df_final, columns=['PC_1', 'PC_2', 'PC_3', 'PC_4'])
df_final.head()
```

### KMEANS CLUSTERING

```python
df_final
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#Elbow method
sse = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df_final)
    sse.append(kmeans.inertia_)
plt.plot(range(2, 11), sse)
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.show()
```

```python
#Silhouette Score
sil = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df_final)
    sil.append(silhouette_score(df_final, kmeans.labels_))
plt.plot(range(2, 11), sil)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score')
plt.show()
```

```python
#K = 
kmeans = KMeans(n_clusters=4)
kmeans.fit(df_final)
df_final['Cluster Assigned'] = kmeans.labels_
df_final
```

```python
df_final.info()
```

```python
ax = df_final['Cluster Assigned'].value_counts().plot(kind='bar')
plt.xlabel('Cluster Assigned')
plt.ylabel('Count')

for p in ax.patches:
    height = p.get_height()
    ax.text(x = p.get_x()+(p.get_width()/2),y = height//2, s = height, ha = 'center')

plt.show()
```
